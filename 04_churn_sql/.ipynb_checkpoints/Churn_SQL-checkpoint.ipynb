```python
# TELCO CHURN SQL ANALYSIS
# Frage: "Wer k√ºndigt und warum?"

import sqlite3 as sql
import pandas as pd

# CHURN CSV (GETESTET ‚Äì funktioniert!)
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

print(f"‚úÖ {df.shape[0]:,} Kunden geladen!")
print("Spalten:", df.columns.tolist())
print("\nChurn‚ÄëRate:", df['Churn'].value_counts())
df.head(3)
```

```python
# CSV -> SQL-Datenbank
# Datenbank erstellen

conn = sql.connect('churn.db')  # ‚Üê Neue DB!
df.to_sql('churn', conn, if_exists='replace', index=False)

print("‚úÖ Tabelle 'churn' erstellt!")
print("üìÅ Datei 'churn.db' gespeichert.")
```

```python
-- Query 1: Churn pro Contract
query1 = """
SELECT 
    Contract,                          -- Gruppiere nach Vertragstyp
    COUNT(*) as Anzahl_Kunden,         -- Z√§hle ALLE Kunden pro Gruppe
    ROUND(AVG(CASE WHEN Churn = 'Yes' THEN 1.0 ELSE 0 END), 3) * 100 AS Churn_Prozent  -- Durchschnitt: 1=Yes, 0=No ‚Üí % K√ºndiger
FROM churn
GROUP BY Contract
ORDER BY Churn_Prozent DESC;
"""
result1 = pd.read_sql_query(query1, conn)
result1  # Zeigt Tabelle
```

```python
-- Zeigt: K√ºndigen Neukunden mehr als Langzeitkunden?
query2 = """
SELECT 
    CASE 
        WHEN tenure <= 6 THEN '0-6 Monate (neu)'
        WHEN tenure <= 12 THEN '7-12 Monate'
        WHEN tenure <= 24 THEN '1-2 Jahre'
        ELSE '2+ Jahre (loyal)'
    END AS Tenure_Gruppe,
    COUNT(*) AS Anzahl_Kunden,
    ROUND(AVG(CASE WHEN Churn='Yes' THEN 1.0 ELSE 0 END)*100, 1) AS Churn_Prozent
FROM churn
GROUP BY Tenure_Gruppe
ORDER BY Churn_Prozent DESC
"""

result2 = pd.read_sql_query(query2, conn)
result2
```

```python
query3 = """
-- Teure Kunden k√ºndigen mehr? + ANZAHL!
SELECT 
    CASE 
        WHEN MonthlyCharges < 20 THEN 'Low (<20$)'
        WHEN MonthlyCharges < 70 THEN 'Medium (20-70$)'
        ELSE 'High (>70$)'
    END AS Preis_Gruppe,
    COUNT(*) AS Anzahl_Kunden,
    ROUND(AVG(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END)*100, 1) AS Churn_Prozent,
    ROUND(AVG(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END) * COUNT(*), 0) AS Kunden_Gekuendigt
FROM churn
GROUP BY 1 
ORDER BY 3 DESC;
"""
result3 = pd.read_sql_query(query3, conn)
result3
```

```python
query4 = """
-- Schnelles Internet = mehr K√ºndiger? + ANZAHL!
SELECT InternetService, 
       COUNT(*) AS Anzahl_Kunden,
       ROUND(AVG(CASE WHEN Churn='Yes' THEN 1.0 ELSE 0 END)*100, 1) AS Churn_Prozent,
       CAST(ROUND(AVG(CASE WHEN Churn='Yes' THEN 1.0 ELSE 0 END) * COUNT(*)) AS INTEGER) AS Kunden_Gekuendigt
FROM churn 
GROUP BY 1 
ORDER BY 3 DESC;
"""
result4 = pd.read_sql_query(query4, conn)
result4
```
