# Amazon Sales Analysis
Dieses Projekt analysiert Verkaufsdaten von Amazon. Ziel ist es:
- Umsatztrends zu erkennen
- Top-Produkte und Kategorien zu identifizieren
- Regionale Unterschiede zu visualisieren
- Ein professionelles Portfolio-Projekt für Data Analysis zu erstellen

```python
# Importieren der Bibliotheken
import pandas as pd  # Pandas ist eine Bibliothek für Datenanalyse. Sie erleichtert das Arbeiten mit Tabellen, wie Excel-Dateien. Wir importieren sie als 'pd', damit wir später nicht immer 'pandas' ausschreiben müssen.
import matplotlib.pyplot as plt  # Matplotlib ist eine Bibliothek, um Diagramme zu erstellen. Wir importieren den Teil 'pyplot' als 'plt', das ist eine gängige Abkürzung.
import seaborn as sns  # Seaborn baut auf Matplotlib auf und macht es einfacher, schöne Diagramme zu erstellen.

# Optionale Einstellung für schöne Diagramme
sns.set(style="whitegrid")  # Seaborn Style
```

```python
# CSV-Datei einlesen
df = pd.read_csv('amazon.csv')  # 'pd.read_csv' liest eine CSV-Datei ein. Wir speichern sie in 'df', das steht für DataFrame – eine Tabelle wie in Excel.
df.head()  # Zeigt die ersten 5 Zeilen der Tabelle an, damit wir einen Überblick bekommen, wie die Daten aussehen.
```

```python
# Überblick über die Daten
df.info()  # Zeigt Spaltennamen, Datentypen und wie viele Werte pro Spalte fehlen. So wissen wir, ob wir Daten bereinigen müssen.
```

```python
df.describe()  # Gibt statistische Zusammenfassungen der numerischen Spalten: Mittelwert, Min, Max, Quartile. Praktisch für ersten Überblick.
```

```python
df.isnull().sum()  # Prüft, wie viele fehlende Werte pro Spalte existieren. Sehr wichtig, bevor wir Analysen starten.
```

```python
# Währungszeichen und Kommas entfernen von String zu Zahl
df['discounted_price'] = pd.to_numeric(df['discounted_price'].astype(str).str.replace('₹','').str.replace(',',''), errors='coerce')
```

```python
# Umsatz pro Kategorie berechnen
category_price = df.groupby('category')['discounted_price'].mean().sort_values(ascending=False)
category_price
```

```python
# Balkendiagramm: Umsatz pro Kategorie
top_categories = category_price.head(10)

plt.figure(figsize=(12,8))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Durchschnittlicher Preis – Top 10 Kategorien')
plt.ylabel('Preis in ₹')
plt.xlabel('Kategorie')

plt.xticks(rotation=45, ha='right')  # ha='right' sorgt dafür, dass die Labels rechtsbündig stehen
plt.tight_layout()  # passt Layout an, damit nichts abgeschnitten wird

plt.show()
```

```python
# Rating in float, falls noch String - Durchschnittliche Bewertung pro Kategorie
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Durchschnittliche Bewertung pro Kategorie berechnen
category_ratings = df.groupby('category')['rating'].mean().sort_values(ascending=False)
top_categories_rating = category_ratings.head(10)

# Balkendiagramm erstellen
plt.figure(figsize=(12,8))
sns.barplot(x=top_categories_rating.index, y=top_categories_rating.values)
plt.title('Durchschnittliche Bewertung – Top 10 Kategorien')
plt.ylabel('Rating')
plt.xlabel('Kategorie')
plt.ylim(3, 5)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top10 Produkte nach Rating
top_products = df.sort_values(by='rating', ascending=False).head(10)
top_products[['product_name', 'rating', 'rating_count', 'discounted_price']]
```

```python
# TOP 3 Werte fürs Fazit extrahieren
print("TEUERSTE KATEGORIEN:")
print(category_price.head(3))

print("\n BESTE RATINGS:")
print(top_categories_rating.head(3))

print("\n TOP PRODUKTE:")
print(top_products[['product_name', 'rating', 'discounted_price']].head(3))
```
